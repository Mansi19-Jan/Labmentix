## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%b-%y')

# Extract Year and Month for analysis
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
````

````python
# Exclude 'Date' from numerical computations
numeric_cols = ['Open', 'High', 'Low', 'Close']

# Compute IQR for numerical columns only
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Filter out outliers using IQR
df = df[~((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

# Save the cleaned dataset
cleaned_file_path = "cleaned_YesBank_StockPrices.csv"
df.to_csv(cleaned_file_path, index=False)

# Display final dataset overview
df.info()
df.head()
````

### What all manipulations have you done and insights you found?

### Data Preprocessing & Cleaning Steps:

**Converted 'Date' Column to Datetime Format:**

The 'Date' column was originally in MMM-YY format (e.g., Jul-05) and was converted to YYYY-MM-DD using pd.to_datetime().
This allows for easy extraction of time-based insights.

**Extracted Year and Month for Analysis:**

Created new columns Year and Month to enable trend analysis over different time periods.

**Handled Outliers Using IQR (Interquartile Range) Method:**

1. *Calculated Q1 (25th percentile) and Q3 (75th percentile) for numerical columns (Open, High, Low, Close).*
2. *Computed the IQR (Q3 - Q1) and removed data points that fell outside the acceptable range (1.5 * IQR).*
3. *This filtered extreme fluctuations and ensured meaningful trend analysis.*
4. *Dataset reduced from 185 to 173 rows, indicating removal of 12 extreme outliers.*
   
**Saved the Cleaned Dataset:**

*The cleaned dataset was stored as cleaned_YesBank_StockPrices.csv for further analysis.*

Final Dataset Overview:

1. *The cleaned dataset contains 173 rows and 7 columns (Date, Open, High, Low, Close, Year, Month).*
2. *No missing values or duplicates exist in the cleaned dataset.*

#### Chart - 6

````python
# Chart - 6 visualization code

# Calculate Percentage Change in Closing Price
df['Price_Change_%'] = df['Close'].pct_change() * 100

# Identify Sudden Spikes & Dips (Threshold: ±10% Change)
shock_events = df[(df['Price_Change_%'] > 10) | (df['Price_Change_%'] < -10)]

# Print the identified sudden shocks
print("Major Sudden Price Changes (±10% change):")
print(shock_events[['Date', 'Close', 'Price_Change_%']])

# Plot Price Changes Over Time
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Price_Change_%'], label="Daily % Change", color='gray', alpha=0.7)

# Highlight Sudden Spikes & Dips
plt.scatter(shock_events['Date'], shock_events['Price_Change_%'], color='red', label="Major Shocks", s=100)

# Add Labels & Title
plt.axhline(y=10, color='green', linestyle='--', label="10% Threshold")
plt.axhline(y=-10, color='red', linestyle='--', label="-10% Threshold")
plt.xlabel("Year", fontsize=12)
plt.ylabel("Price Change (%)", fontsize=12)
plt.title("Yes Bank Stock: Sudden Price Shocks", fontsize=14)
plt.grid()
plt.legend()
plt.show()
````

#### Stock Price Trends visualization with Simple Moving Average (SMA) and Exponential Moving Average (EMA) 

````python
# Chart - 7 visualization code

# Calculate Moving Averages
df['SMA_50'] = df['Close'].rolling(window=50).mean()  # 50-day Simple Moving Average
df['EMA_20'] = df['Close'].ewm(span=20, adjust=False).mean()  # 20-day Exponential Moving Average

# Plot the data
plt.figure(figsize=(12,6))
plt.plot(df['Close'], label='Closing Price', color='blue', alpha=0.6)
plt.plot(df['SMA_50'], label='50-Day SMA', color='red', linestyle='dashed')
plt.plot(df['EMA_20'], label='20-Day EMA', color='green')

# Title and labels
plt.title("Yes Bank Stock Price Trends with SMA & EMA")
plt.xlabel("Date")
plt.ylabel("Stock Price (INR)")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
````

##### 1. Why did you pick the specific chart?

* This chart effectively captures long-term and short-term price trends using Simple Moving Average (SMA) and Exponential Moving Average (EMA).
* SMA (50-day) helps to identify long-term trends, while EMA (20-day) provides insight into short-term momentum shifts.
* The visualization helps detect bullish or bearish signals in stock price movements.

##### 2. What is/are the insight(s) found from the chart?

* If EMA (green) crosses below SMA (red) → Bearish signal (downtrend).
* SMA smooths out long-term trends, while EMA reacts faster to price movements.

*Key Findings:*
