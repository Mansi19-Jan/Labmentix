## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Handling missing values
# 1. Drop columns with high missing values (if necessary)
threshold = 0.5  # If more than 50% data is missing, drop the column
df = df.dropna(thresh=int(threshold * len(df)), axis=1)

# 2. Filling missing values
# For numerical columns, fill with mean
data.update(data.select_dtypes(include=[np.number]).apply(lambda x: x.fillna(x.mean())))

# For categorical columns, fill with mode
data.update(data.select_dtypes(include=[object]).apply(lambda x: x.fillna(x.mode()[0])))

# Removing duplicates
data.drop_duplicates(inplace=True)

# Resetting index after cleaning
data.reset_index(drop=True, inplace=True)

````

````python
# Visualizing Outliers using Boxplot
plt.figure(figsize=(15, 8))
for column in data.select_dtypes(include=[np.number]).columns:
    plt.figure()
    sns.boxplot(x=data[column])
    plt.title(f'Boxplot of {column}')
    plt.show()

# Detecting Outliers using Z-Score
z_scores = np.abs((data.select_dtypes(include=[np.number]) - data.select_dtypes(include=[np.number]).mean()) / data.select_dtypes(include=[np.number]).std())

# Setting threshold for Z-score (e.g., 3)
outliers = (z_scores > 3).any(axis=1)

# Removing Outliers
data = data[~outliers]

````

````python
# Final Dataset Info
print("\nCleaned Dataset Info:")
print(data.info())

# Save the cleaned data
df.to_csv('Cleaned_Customer_Support_Data.csv', index=False)
````
