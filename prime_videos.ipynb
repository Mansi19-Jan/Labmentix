## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

#Drop duplicates
titles_df = df1.drop_duplicates()
credits_df = df2.drop_duplicates()

common_cols = set(titles_df.columns).intersection(set(credits_df.columns))
print("Common columns for merge:", common_cols)

# Merge datasets (adjust merge key if different)
merge_key = 'id' if 'id' in common_cols else list(common_cols)[0]  # fallback to first common column
merged_df = pd.merge(titles_df, credits_df, on=merge_key, how='left')

#Convert data types
if 'release_year' in merged_df.columns:
    merged_df['release_year'] = pd.to_numeric(merged_df['release_year'], errors='coerce')

if 'duration' in merged_df.columns:
     merged_df['duration'] = merged_df['duration'].astype(str).str.extract(r'(\d+)').astype(float)

print("✅ Data wrangling completed. Cleaned dataset shape:", merged_df.shape)
````

````python
merged_df.head()
````

### What all manipulations have you done and insights you found?

Data wrangling, also known as data cleaning or data preprocessing, is a crucial step in any data analysis project. It involves transforming raw data into a clean and usable format for further analysis and visualization. In this project, data wrangling was essential to ensure the accuracy and reliability of insights.

1. Removing Duplicates
* To prevent biased analysis due to repeated entries, we used the drop_duplicates() method to remove any duplicate rows from both datasets — titles_df and credits_df.

2. Merging Datasets
* We had two datasets — one containing title information (like movie/series name, release year, runtime, ratings) and the other containing credits (like cast and crew). We merged them using a left join on the common column id, ensuring that all titles were preserved even if some had missing cast data.

3. Handling Missing Values
* We analyzed and visualized missing values in each column using bar plots. Key columns like seasons, age_certification, character, and tmdb_score had significant missing data. Based on the context, we plan to either drop, fill, or ignore them depending on their impact on the analysis.

4. Data Type Conversion
 Some columns such as release_year and duration were not in the appropriate numeric format. We converted:
 * release_year from string to numeric, coercing invalid entries.
 * duration (like "120 min") to numeric by extracting only the digits and converting them to float.

5. Final Cleaned Dataset
After cleaning, merging, and converting data types, our final dataset had:
* 125,186 rows
* 19 cleaned and meaningful columns
This dataset is now ready for deeper exploration, visualization, and statistical analysis.

Why Data Wrangling is Important
* Improves data quality by removing errors, nulls, and inconsistencies.
* Prepares data for analysis and visualization by converting it into the right format.
* Avoids biased or incorrect conclusions due to messy or inaccurate data.
* Ensures smooth performance in machine learning models, if used later.
